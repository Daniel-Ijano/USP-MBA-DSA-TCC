Data Queries

```python
import mysql.connector
import pandas as pd

try:
    conn = mysql.connector.connect(
            #use_pure=True,
            host='localhost',
            user='root',
            passwd='1234',
            database="news"
        )

    query = (
    "SELECT section, title "
    "FROM tb_news "
    "WHERE section NOT IN ('Política') "
    "ORDER BY title DESC "
    "LIMIT 10"
    )

    result_dataFrame = pd.read_sql(query,conn)
    conn.close()

except Exception as e:
    conn.close()
    print(str(e))
```

```python
#VADER EXAMPLE
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer #usando um corpora pronto
#nltk.download()

vader = SentimentIntensityAnalyzer()

f = lambda title: vader.polarity_scores(title)['compound']
result_dataFrame['compound'] = result_dataFrame['title'].apply(f)

result_dataFrame
```

```python
#TOKENIZE EXAMPLE
import nltk    
from nltk import tokenize

result_dataFrame['tokenized_sents'] = result_dataFrame.apply(lambda row: nltk.word_tokenize(row['title']), axis=1)

result_dataFrame
```

```python
#NGRAM EXAMPLE
from nltk.tokenize import word_tokenize
from nltk.util import ngrams

def get_ngrams(text, n ):
    n_grams = ngrams(word_tokenize(text), n)
    return [ ' '.join(grams) for grams in n_grams]

result_dataFrame['ngram_sents'] = result_dataFrame.apply(lambda row: get_ngrams(row['title'], 2), axis=1)

result_dataFrame
```
